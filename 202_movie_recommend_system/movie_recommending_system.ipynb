```python
def cast_converter(obj):
    l = []
    count = 0
    for i in ast.literal_eval(obj):
        if count < 5:
            l.append(i['name'])
            count += 1
        else:
            break
    return l
```

```python
def crew_converter(obj):
    if isinstance(obj, str):
        try:
            obj = ast.literal_eval(obj)
        except:
            return []

    # Step 2: Filter only 'Director' (case-insensitive)
    directors = []
    for i in obj:
        if isinstance(i, dict) and i.get('job', '').lower() == 'director':
            directors.append(i['name'])

    return directors
```

```python
movies['crew'] = movies['crew'].apply(crew_converter)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# steaming is like assigning same word for past,present,future like this
def stem(text):
  y = []
  for i in text.split(): # to string
    y.append(ps.stem(i)) # to list
  return " ".join(y)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
# euclidian distance is not a good measure in higher dimensions
# cosine distance is optimal choice
# distance is inversly proportional to similarilty
```

```python
from sklearn.metrics.pairwise import cosine_similarity  # directly we can import similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape # each movies with all other movies
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie):
  movie_index = df[df['title']==movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key= lambda x:x[1])[1:6]

  for i in movies_list:
    print(df.iloc[i[0]].title)
```

```python
recommend('The Dark Knight')
```
